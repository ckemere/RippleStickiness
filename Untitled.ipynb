```python
import numpy as np
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

Path = mpath.Path
fig, ax = plt.subplots()

pathdata = [
    (Path.MOVETO, (100, 200)),
    (Path.LINETO, (100, 100)),
    (Path.LINETO, (150, 100)),
    (Path.LINETO, (200, 100)),
    (Path.LINETO, (200, 200)),
    (Path.MOVETO, (150, 200)),
    (Path.LINETO, (150, 100)),
    ]

codes, verts = zip(*pathdata)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, facecolor='green', edgecolor='yellow', alpha=0.5)
ax.add_patch(patch)
ax.set_xlim(0,300)
ax.set_ylim(0,300)

plt.show()
```
